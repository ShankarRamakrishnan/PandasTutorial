# Time Series

```
import pandas as pd
import numpy as np
```

Create a simple time pandas Series of length 50 for every second 

```
# Creating a data range
dates = pd.date_range('20200302', periods=500, freq='s')
dates
```

Make the dates as the index of the new series of random numbers

```
ts = pd.Series(np.random.randint(0, 500, len(dates)), dates)
ts
```

What is resampling?

Resampling is necessary when youâ€™re given a data set recorded in some time interval and you want to change the time interval to something else. For example, you could aggregate monthly data into yearly data, or you could upsample hourly data into minute-by-minute data.

```
ts.resample('5min').sum
ts
```

See the time in UTC

```
ts_utc = ts.tz_localize('UTC')
ts_utc
```

Convert to different time zone

```
ts_utc.tz_convert('US/Eastern')
```
